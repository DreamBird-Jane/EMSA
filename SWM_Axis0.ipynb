```python
%matplotlib inline
import pandas as pd
import emsalib as emsa
import matplotlib.pyplot as plt
```

```python
dataset_path = 'data/AliceShort.csv'
header = pd.read_csv(dataset_path, nrows=1, header=None)
header = header.values[0]
alice = pd.read_csv(dataset_path)
```

```python
ts = alice['Timestamp'].values
ts_orig = ts - min(ts)
axis = [alice['Axis' + str(i+1)].values for i in range(3)]

distThres = 0.062
```

```python

swm0 = emsa.SWM(distThres = distThres, minLen = 50, maxLen=51, searchRange=10, searchOrd= 15, filterLen = 5)

(motifs, meanErr, droppedPts, distList) = swm0.process(axis)
```

```python
plt.figure(figsize=(4*1.1, 3*1.1), dpi=160)
motifs.plot() 
plt.grid()
plt.title('Sliding window extracted motifs')
plt.xlabel('Real Time Axis (s)')
plt.ylabel('Acceleration (G)')
plt.savefig('SWM_extracted_motif.png', bbox_inches="tight", pad_inches=0)

plt.figure(figsize=(4*1.1, 3*1.1), dpi=160)
plt.plot(distList, 'o-', markersize=2, color='red', markerfacecolor='blue', markeredgecolor='blue')
plt.axhline(y=distThres, color='black',lw=2)
plt.grid()
#plt.ylim(0.01,0.1)
#plt.xlim(1000, 1200)
plt.title('Distance between template and canditate')
plt.ylabel('Euclidean distance')
plt.xlabel('Iteration')
plt.savefig('SWM_dist_epoch.png', bbox_inches="tight", pad_inches=0)
```
